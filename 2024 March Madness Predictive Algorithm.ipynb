```python
import csv
#need for reader

def adjusted_eff_margin(filename):
  col_index_2 = 2
  team_name = 1
  adjEM_stats = {}

  with open(filename, 'r') as file:
      reader = csv.reader(file)
      next(reader)
      for row in reader:
        team = row[team_name]
        stat = float(row[col_index_2])
        if team in adjEM_stats:
          adjEM_stats[team].append(stat)
        else:
          adjEM_stats[team] = [stat]




  total_sum = 0
  total_count = 0
  for scores in adjEM_stats.values():
    total_sum += sum(scores)
    total_count += len(scores)

  # Calculate the mean
  mean = total_sum / total_count


  processed_adjEM = {}
  for team, stats in adjEM_stats.items():
    mean_team = sum(stats) / len(stats)
    processed_adjEM[team] = (mean_team / mean) * 50
  # print("Adjusted Efficiency Margin:")
  # print(processed_adjEM)
  return processed_adjEM




def offensive_eff(filename):
  #offensive effciency stats are in column 3
  offensive_eff_column = 3
  #team namaes are in column 2
  team_name_column = 2


  off_eff_dict = {}

  with open(filename, 'r') as file:
      reader = csv.reader(file)
      next(reader)
      for offensive_eff_row in reader:
        team = offensive_eff_row[team_name_column]
        #using csv.reader to reference specific columns
        offsenive_eff_stat = float(offensive_eff_row[offensive_eff_column])
        if team in off_eff_dict:
          off_eff_dict[team].append(offsenive_eff_stat)
        else:
          off_eff_dict[team] = [offsenive_eff_stat]



  total_sum = 0
  total_count = 0
  for scores in off_eff_dict.values():
    total_sum += sum(scores)
    total_count += len(scores)

  # Calculate the mean
  mean = total_sum / total_count

  processed_offE = {}
  for team, stats in off_eff_dict.items():
    mean_team = sum(stats) / len(stats)
    processed_offE[team] = (mean_team / mean) * 100
  # print("Offensive Efficiency:")
  # print(processed_offE)
  return processed_offE




def defensive_eff(filename):
  col_index_3 = 3
  team_name = 2
  def_eff = {}
  with open(filename, 'r') as file:
      reader = csv.reader(file)
      next(reader)
      for row in reader:
        team = row[team_name]
        stat = float(row[col_index_3])
        # using reader to be able to reference specific columns
        if team in def_eff:
          def_eff[team].append(stat)
        else:
          def_eff[team] = [stat]

  total_sum = 0
  total_count = 0
  for scores in def_eff.values():
    total_sum += sum(scores)
    total_count += len(scores)

  # mean calculation
  mean = total_sum / total_count

  processed_defE = {}
  for team, scores in def_eff.items():
    processed_defE[team] = (mean / sum(scores)) * 100

  # print("Defensive Efficiency:")
  # print(processed_defE)
  return processed_defE




def strength_of_schedule(filename):
  col_index_2 = 3
  team_name = 1
  SOS_stats = {}
  with open(filename, 'r') as file:
      reader = csv.reader(file)
      next(reader)
      for row in reader:
        team = row[team_name]
        stat = float(row[col_index_2])
        #making the dictionary
        if team in SOS_stats:
          SOS_stats[team].append(stat)
        else:
          SOS_stats[team] = [stat]

  total_sum = 0
  total_count = 0
  for scores in SOS_stats.values():
    total_sum += sum(scores)
    total_count += len(scores)


  # Calculate the mean
  mean = total_sum / total_count

  #weighing the stat
  processed_SOS = {}
  for team, stats in SOS_stats.items():
    mean_team = sum(stats) / len(stats)
    processed_SOS[team] = (mean_team / mean) * 55


  return processed_SOS



def turnovers(filename):
  #turnovers column 3 (index 2)
  turnover_column = 2
  #team names are in column 1
  team_name_column = 1

  turnovers_dict = {}

  with open(filename, 'r') as file:
      reader = csv.reader(file)
      next(reader)
      for turnovers_row in reader:
        team = turnovers_row[team_name_column]
        turnover_stat = float(turnovers_row[turnover_column])
        if team in turnovers_dict:
          turnovers_dict[team].append(turnover_stat)
        else:
          turnovers_dict[team] = [turnover_stat]

  total_sum = 0
  total_count = 0
  for scores in turnovers_dict.values():
    total_sum += sum(scores)
    total_count += len(scores)

  # Calculate the mean
  mean = total_sum / total_count

# weighing the tunrover stat
  processed_turnovers = {}
  for team, scores in turnovers_dict.items():
    processed_turnovers[team] = (mean / sum(scores)) * 80

  return processed_turnovers




def combined_team_scores():
  adjEM_combined = adjusted_eff_margin("SI 106 Project Data - AdjustedEM and SOS (1).csv")
  off_Eff_combined = offensive_eff("SI 106 Project Data - Offensive Efficiency.csv")
  def_Eff_combined = defensive_eff("SI 106 Project Data - Defensive Effiency.csv")
  turnovers_combined = turnovers("SI 106 Project Data - Turnovers.csv")
  sos_combined = strength_of_schedule("SI 106 Project Data - AdjustedEM and SOS (1).csv")

  combined_scores = {}
  for team, score in adjEM_combined.items():
    if team not in combined_scores:
      combined_scores[team] = 0
    combined_scores[team] += float(0.3 * float(score))


  for team, score in off_Eff_combined.items():
    combined_scores[team] += float(0.3 * float(score))


  for team, score in def_Eff_combined.items():
    combined_scores[team] += float(0.3 * float(score))

  for team, score in turnovers_combined.items():
    combined_scores[team] += float(0.05 * float(score))

  for team, score in sos_combined.items():
    combined_scores[team] += float(0.05 * float(score))
  sorted_teams = dict(sorted(combined_scores.items(), key=lambda item: item[1], reverse=True))

  # print(sorted_teams)
  return sorted_teams


# combined_team_scores()

def scores_by_seed(seed_choice, filename):
  seed_column = 0

  team_name_column = 2

  seed_dict = {}

  with open(filename, 'r') as file:
      reader = csv.reader(file)
      next(reader)
      for row in reader:
        team = row[team_name_column]
        seed_dict[team] = row[seed_column]

  seed_choice_teams = []
  for key in seed_dict:
    if int(seed_dict[key]) == seed_choice:
      seed_choice_teams.append(key)

  full_team_scores = combined_team_scores()
  return_dict = {}
  for team in seed_choice_teams:
    return_dict[team] = full_team_scores[team]
  sorted_seeded_teams = dict(sorted(return_dict.items(), key=lambda item: item[1], reverse=True))
  print(f"Ranking of all of the {seed_choice} seeded teams:")

  count = 1
  for key in sorted_seeded_teams:
    print(f"{count}: {key}")
    count += 1

#scores_by_seed(14, "SI 106 Project Data - Offensive Efficiency (6).csv")
def run_program():
  print("Welcome to our March Madness Algorithm!\n\nWe have developed a new system that weighs:\n-Adjusted Efficncy Margin\n-Offensive Efficiency\n-Defenseive Efficiency\n-Strength of Schedule\n-Amount of Turnovers\nThe Program Outputs teams in order of most likely to win the 2024 NCAA Men's Basketball Tournament")
  first_choice = input("Type 1 for a comprehensive ranking of the 68 teams.\nType 2 to see the ranking of all the teams of a particular seed.\nType done to end program.\n")
  while first_choice != "done":
    #try except for making sure input is a number
    try:
      int(first_choice)
    except:
      if first_choice.lower() == "done":
        return
    #we have a try except for bad inputs
    try:
      #choice for the full list
      if int(first_choice) == 1:
        team_rankings = combined_team_scores()
        count = 1
        for key in team_rankings:
          print(f"{count}: {key}, {team_rankings[key]}")
          count += 1
        first_choice = input("Type 1 for a comprehensive ranking of the 68 teams.\nType 2 to see the ranking of all the teams of a particular seed.\nType done to end program.\n")
      #choice for the ranking of a certain seed
      if int(first_choice) == 2:
        seed_choice = input("Please enter a seed (1-16): ")
        #checking for wrong inputs
        while not seed_choice.isdigit() or int(seed_choice) < 1 or int(seed_choice) > 16:
          seed_choice = input("Please enter a seed (1-16): ")
        scores_by_seed(int(seed_choice), "SI 106 Project Data - Offensive Efficiency (6).csv")
        first_choice = input("Type 1 for a comprehensive ranking of the 68 teams.\nType 2 to see the ranking of all the teams of a particular seed.\nType done to end program.\n")
    except:
        first_choice = input("Type 1 for a comprehensive ranking of the 68 teams.\nType 2 to see the ranking of all the teams of a particular seed.\nType done to end program.\n")

run_program()
```
